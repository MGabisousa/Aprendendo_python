## Descrição dos dados

A tabela `"SIGAA.csv"` é uma tabela com informações de acesso dos alunos no sistema SIGAA.

- A primeira coluna conta com as datas de acesso (o primeiro dia é um domingo);
- A segunda coluna se refere à matrícula do aluno que acessou o sistema (os quatro primeiros dígitos da matrícula diz respeito ao ano de ingresso do aluno na universidade);
- A terceira coluna é o e-mail institucional (o email segue o padrão sobrenome.nome@centro.ufpb.br);
- A quarta coluna apresenta o CRA do aluno que fez acesso ao sistema.

Seu trabalho é seguir as instruções das células abaixo para fazer os processamentos dos dados.
